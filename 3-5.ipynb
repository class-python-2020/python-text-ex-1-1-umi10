```python
while　条件式:
    繰り返したい処理
```

```python
a=1
while a<=10:
    print(a)
    a +=1
```

```python
a,b = 0,1
while b < 20:
    print(b)
    a,b = b, a+b
```

```python
a = 1
while True:
    if a%2==0:
        print(a)
    a +=1
```

```python
a=1
while a<10:
    if a%2==0:
        print(a)
    a+=1
```

```python
a = 1
while True:
    if a %2 == 0:
        print(a)
    a+=1
    if a>=10:
        break
```

```python
a = 0
while a < 100:
    if a > 10:
        break
    a += 2
print(a)
```
